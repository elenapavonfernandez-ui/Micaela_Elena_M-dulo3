```python
import numpy as np

# 1. Crear el array 1D
array_1d = np.random.randint(10, 100, 50) 
print(f"Array original (primeros 10): {array_1d[:10]}...")

# 2. Generar y filtrar los índices
indices = np.arange(len(array_1d))
mascara_indices = (indices % 5 == 0)

# 3. Acceder a los elementos en esas posiciones
elementos_seleccionados = array_1d[mascara_indices]

print("\nÍndices seleccionados: [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]")
print(f"Elementos en esas posiciones: {elementos_seleccionados}")
```
