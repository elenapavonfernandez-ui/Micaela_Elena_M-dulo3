- country: País.
- density: Densidad de población (personas por km²).
- abbreviation: Abreviatura o código corto del país.
- agricultural_land: Porcentaje de tierra destinada a agricultura.
- land_area: Superficie terrestre total del país (km²).
- armed_forces_size: Tamaño de las fuerzas armadas (número de militares).
- birth_rate: Tasa de natalidad (nacimientos por cada 1.000 habitantes).
- calling_code: Código telefónico internacional del país.
- capital/major_city: Capital o ciudad principal.
- co2-emissions: Emisiones de CO₂ del país.
- cpi: Índice de Precios al Consumidor (inflación).
- cpi_change: Cambio anual del IPC.
- currency-code: Código de la moneda oficial.
- fertility_rate: Tasa de fertilidad (número medio de hijos por mujer).
- forested_area: Porcentaje del territorio cubierto por bosques.
- gasoline_price: Precio de la gasolina.
- gdp: Producto Interno Bruto.
- gross_primary_education_enrollment: Tasa bruta de matrícula en educación primaria.
- gross_tertiary_education_enrollment: Tasa bruta de matrícula en educación terciaria/universitaria.
- infant_mortality: Mortalidad infantil (muertes por cada 1.000 nacidos vivos).
- largest_city: Ciudad más grande del país.
- life_expectancy: Esperanza de vida.
- maternal_mortality_ratio: Mortalidad materna (por cada 100.000 nacimientos).
- minimum_wage: Salario mínimo del país.
- official_language: Idioma oficial.
- out_of_pocket_health_expenditure: Gasto sanitario de bolsillo por parte de los ciudadanos.
- physicians_per_thousand: Médicos por cada 1.000 habitantes.
- population: Población total.
- population_labor_force_participation: Participación en la fuerza laboral (% de población activa).
- tax_revenue: Ingresos provenientes de impuestos.
- total_tax_rate: Tasa total de impuestos.
- unemployment_rate: Tasa de desempleo.
- urban_population: Población urbana (en porcentaje o número según dataset).
- latitude: Latitud geográfica del país.
- longitude: Longitud geográfica del país.

¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

```python
df.info()
```

```python
sns.countplot(y = "continent", 
              data = df, 
              color = "c"); #Esta no nos da los porcentajes
```

```python
df_continent = df.groupby("continent")["country"].count().reset_index()
df_continent
```

```python
plt.pie(data=df.groupby("job")["marital"].count().reset_index(), x="marital", labels="id",autopct= );
```

```python
plt.pie(x= "country", labels= "continent",
        data = df_continent, 
        autopct=  '%1.1f%%', 
        textprops={'fontsize': 8}, 
        startangle=90);
```

¿Cuáles son los 10 países con la mayor área forestal?

```python
df_forestal = df[["forested_area", "country"]].reset_index().sort_values(by="forested_area", ascending=False).head(10)
df_forestal
```

```python
fig, axes = plt.subplots(figsize = (10,5))


# creamos nuestro primer barplot. 📌 LO QUE NOS MUESTRA ES LA MEDIA DE EDAD EN FUNCIÓN DEL ESTADO CIVIL
sns.barplot(x = "forested_area", 
            y = "country", 
            data = df_forestal, 
            palette = "magma")
;
```
